# object 값들 encoding 해주기

```python
# label encoding
from sklearn.preprocessing import LabelEncoder

lab = LabelEncoder()
```

```python
bike_df = bike.copy()

bike_df['seller_type']=lab.fit_transform(bike_drop_ex['seller_type'])
bike_df['owner']=lab.fit_transform(bike_drop_ex['owner'])
bike_df['name']=lab.fit_transform(bike_drop_ex['name'])
```

```python
bike_df
```

```python
bike_df.seller_type.value_counts()
```

> 1 : Individual, 0: Dealer

```python
bike_df.owner.value_counts()
```

> 0 : 1st owner, 1: 2nd owner, 2: 3rd owner, 3: 4th owner

```python
bike_df.name = bike_df.name/bike_df.name.max()
bike_df.selling_price = bike_df.selling_price/bike_df.selling_price.max()
bike_df.year = bike_df.year/bike_df.year.max()
bike_df.km_driven = bike_df.km_driven/bike_df.km_driven.max()
bike_df.ex_showroom_price = bike_df.ex_showroom_price/bike_df.ex_showroom_price.max()
```

# Train, Test

```python
bike_feature = bike_df.drop('selling_price', axis=1)
bike_target = bike_df.selling_price
```

```python
bike_feature.shape, bike_target.shape
```

```python
from sklearn.model_selection import train_test_split, cross_validate
```

```python
train_input, test_input, train_target, test_target = train_test_split(
    bike_feature, bike_target, test_size=0.2
)
```

```python
# 학습, 검증 데이터셋 분리
sub_input, val_input, sub_target, val_target = train_test_split(
    train_input, train_target, test_size=0.2
)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
# 학습
lr=LinearRegression()
model=lr.fit(train_input,train_target)

# 검증
print('train score :', model.score(train_input, train_target))
print('sub score :', model.score(sub_input, sub_target))
print('val score :', model.score(val_input, val_target))

print('predict :', model.predict(test_input)[0:10])
```
