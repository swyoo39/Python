# featureÏùÑ ÏàòÏπòÍ∞íÏùò 11 + 2Í∞ú Ïª¨ÎüºÎì§Î°ú, targetÏùÑ Style Ïª¨ÎüºÏúºÎ°ú

```python
bFeature = beer_df_new.loc[:,'Astringency':'Malty']
bFeature['ABV'] = beer_df_new.ABV
# bFeature['Ave Rating'] = beer_df_new['Ave Rating']
bTarget = beer_df_new.Style
```

```python
bTarget.shape, bFeature.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_input, test_input, train_target, test_target = train_test_split(
    bFeature, bTarget, test_size=0.2
)
```

```python
train_input.shape, train_target.shape
```

```python
# train ÏÖãÏúºÎ°ú ÌõàÎ†®Í≥º Í≤ÄÏ¶ù ÏÖãÏùÑ 8:2Î°ú Î∂ÑÎ¶¨ÌïúÎã§.
sub_input, val_input, sub_target, val_target = train_test_split(
    train_input, train_target, test_size=0.2
)
```

```python
train_input.shape, sub_input.shape, val_input.shape
```

# ÌïôÏäµÍ≥º Í≤ÄÏ¶ù

```python
# KNN
from sklearn.neighbors import KNeighborsClassifier

nn = KNeighborsClassifier(n_neighbors=3) # kÍ∞í ÎîîÌè¥Ìä∏ 5, Ï¥òÏ¥òÌï†Îïê ÏûëÍ≤å Ï§ÑÏù¥Ïûê : hyper parameter
nn.fit(sub_input, sub_target)
print('train score :', nn.score(sub_input, sub_target))
print('valid score :', nn.score(val_input, val_target))
```

```python
# Decision Tree
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier().fit(sub_input, sub_target)
print('train score :', dt.score(sub_input, sub_target))
print('valid score :', dt.score(val_input, val_target))
```

```python
# Random Forest
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier().fit(sub_input, sub_target)
print('train score :', rf.score(sub_input, sub_target))
print('valid score :', rf.score(val_input, val_target))
```

```python
# AdaBoost
from sklearn.ensemble import AdaBoostClassifier

ab = AdaBoostClassifier()
ab.fit(sub_input, sub_target)
print('train score :', ab.score(sub_input, sub_target))
print('valid score :', ab.score(val_input, val_target))
```

```python
# Neural Net : Ïù∏Í≥µÏã†Í≤ΩÎßù
from sklearn.neural_network import MLPClassifier

net = MLPClassifier(max_iter=1000)
net.fit(sub_input, sub_target)
print('train score :', net.score(sub_input, sub_target))
print('valid score :', net.score(val_input, val_target))
```

```python
# Naive Bayes
from sklearn.naive_bayes import GaussianNB

nb = GaussianNB() # hyper parameter Îß§Ïö∞ ÎßéÎã§. ÌïÑÏöîÌï† Îïå Í≤ÄÏÉâ
nb.fit(sub_input, sub_target)
print('train score :', nb.score(sub_input, sub_target))
print('valid score :', nb.score(val_input, val_target))
print('test score :', nb.score(test_input, test_target))
```

> Naive Bayes Î™®Îç∏Ïù¥ Í∞ÄÏû• Ï†ÅÌï©ÌïòÎã§.

```python
# Ï†ïÏò§Ìëú
from sklearn import metrics

print(metrics.confusion_matrix(test_target, nb.predict(test_input)))
```

```python
# Îã®Ïñ¥Î•º Ïà´ÏûêÎ°ú Ïù∏ÏΩîÎî©
paragraphs = beer_df.Description.to_list()
paragraphs[0:5]
```

```python
keras.utils.set_random_seed(1)
```

```python
# one hot encoding
encoded_paragraphs = [keras.preprocessing.text.one_hot(paragraph, vocab_size) for paragraph in paragraphs]
encoded_paragraphs
```

```python
# Îç∞Ïù¥ÌÑ∞ÏóêÏÑú Í∞ÄÏû• Í∏¥ Î¨∏Ïû•Ïùò Îã®Ïñ¥ Ïàò ÌôïÏù∏
max_length = 0
for row in beer_df.Description:
    if len(row.split(' ')) > max_length:
        max_length = len(row.split(' '))
print(max_length)
```

```python
# Î¨∏Ïû•ÎßàÎã§ Îã®Ïñ¥ ÏàòÍ∞Ä Îã§Î•¥ÎØÄÎ°ú sequence paddingÏùÑ ÎÑ£Ïñ¥ÏÑú Î¨∏Ïû•Ïùò Í∏∏Ïù¥Î•º ÎèôÏùºÌïòÍ≤å ÎßåÎì§Í∏∞
padded_paragraphs_encoding = keras.preprocessing.sequence.pad_sequences(encoded_paragraphs, maxlen=max_length, padding='post')
padded_paragraphs_encoding
```
