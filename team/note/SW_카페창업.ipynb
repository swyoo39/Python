```python
ss = StandardScaler().fit_transform(feature_oh)
feature_oh_ss = pd.DataFrame(ss, columns=feature_oh.columns)
feature_oh_ss
```

```python
pca_df_10
```

```python
# 그래프 그리기
plt.plot(pca.explained_variance_ratio_)
plt.show()
```

```python
pca = PCA(n_components=5) # n_components : 4개만큼의 feature을 선택
pca_transform = pca.fit_transform(feature_oh_ss)
pca_df_5 = pd.DataFrame(pca_transform)
pca_df_5
```

# train, test

```python
from sklearn.model_selection import train_test_split
```

```python
train_input, test_input, train_target, test_target = train_test_split(
    feature_oh_scaled, target, test_size=0.2
)
```

```python
train_input.shape, train_target.shape, test_input.shape, test_target.shape
```

```python
sub_input, val_input, sub_target, val_target = train_test_split(
    train_input, train_target, test_size=0.2
)
```

```python
sub_input.shape, sub_target.shape, val_input.shape, val_target.shape
```

---
# 학습, 검증

```python
# KNN
from sklearn.neighbors import KNeighborsClassifier

nn_clf = KNeighborsClassifier(n_neighbors=3) # k값 디폴트 5, 촘촘할땐 작게 줄이자 : hyper parameter
nn_clf.fit(sub_input, sub_target)
print('sub 정답률 :', nn_clf.score(sub_input, sub_target))
print('val 정답률 :', nn_clf.score(val_input, val_target))
```

```python
# Decision Tree 
from sklearn.tree import DecisionTreeClassifier

dt_clf = DecisionTreeClassifier()
dt_clf.fit(sub_input, sub_target)
print('sub 정답률 :', dt_clf.score(sub_input, sub_target))
print('val 정답률 :', dt_clf.score(val_input, val_target))
```

```python
# Random Forest 
from sklearn.ensemble import RandomForestClassifier

rf_clf = RandomForestClassifier()
rf_clf.fit(sub_input, sub_target)
print('sub 정답률 :', rf_clf.score(sub_input, sub_target))
print('val 정답률 :', rf_clf.score(val_input, val_target))
```

```python
# AdaBoost
from sklearn.ensemble import AdaBoostClassifier

ab_clf = AdaBoostClassifier()
ab_clf.fit(sub_input, sub_target)
print('sub 정답률 :', ab_clf.score(sub_input, sub_target))
print('val 정답률 :', ab_clf.score(val_input, val_target))
```

```python
# Neural Net : 인공신경망
from sklearn.neural_network import MLPClassifier

net_clf = MLPClassifier()
net_clf.fit(sub_input, sub_target)
print('sub 정답률 :', net_clf.score(sub_input, sub_target))
print('val 정답률 :', net_clf.score(val_input, val_target))
```

```python
# Naive Bayes
from sklearn.naive_bayes import GaussianNB

nb_clf = GaussianNB() # hyper parameter 매우 많다. 필요할 때 검색
nb_clf.fit(sub_input, sub_target)
print('sub 정답률 :', nb_clf.score(sub_input, sub_target))
print('val 정답률 :', nb_clf.score(val_input, val_target))
```
