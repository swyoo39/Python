```python
import pandas as pd
```

```python
df_gold = pd.read_csv('../data/gold.csv')
df_gold.head()
```

- date : 금시장 시세 날짜
- open : 개장 금값
- high : 금값 최고점
- low : 금값 최저점
- close : 폐장 금값
- volume : 거래량
- currency : 화폐

```python
# 모든 거래가 USD 기준이다.
df_gold.drop('Currency', axis=1, inplace=True)
```

```python
df_gold.head()
```

```python
df_gold.tail()
```

```python
import numpy as np
```

```python
np.arange(2000, 2023, 1)
```

```python
Year = []
for i in range(len(df_gold)):
    for j in np.arange(2000, 2023, 1):
        if df_gold.Date[i][0:4]=='%d'%j:
            Year.append('%d'%j)
df_gold['Year'] = Year
df_gold.head()
```

```python
df_gold.tail()
```

```python
np.arange(1, 32, 1)
```

```python
Month = []
for i in range(len(df_gold)):
    for j in np.arange(1, 32, 1):
        if df_gold.Date[i][5:7]=='%.2d'%j:
            Month.append('%.2d'%j)
df_gold['Month'] = Month
df_gold.head()
```

```python
df_gold.tail()
```

```python
df_gold['YM'] = df_gold.Year + '-' + df_gold.Month
df_gold.head()
```

```python
### YM 을 기준으로 groupby 하기
df_gold.groupby('YM').sum()
```

> rows 수가 너무 적다.

```python
df_gold_date = df_gold.set_index('Date').drop(['Year','Month','YM'], axis=1)
df_gold_date
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(df_gold_date.Open)
plt.show()
```

```python
np.corrcoef(df_gold_date.High, df_gold_date.Volume)
```
