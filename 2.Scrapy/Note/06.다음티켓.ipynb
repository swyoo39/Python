### 다음 티켓 가져오나?

```python
#mainContent > div > div.box_boxoffice > ol > li:nth-child(1) > div > div.thumb_cont > strong > a
```

```python
from bs4 import BeautifulSoup as BS
from urllib.request import urlopen
import pandas as pd
```

```python
url = 'https://movie.daum.net/ranking/boxoffice/weekly'
page = urlopen(url)
soup = BS(page, 'html.parser')
```

```python
li_list = soup.select('#mainContent > div > div.box_boxoffice > ol > li > div > div.thumb_cont > strong > a')
print(li_list)
```

```python
# data frame 만들기
count = 0
titleList = [] # 영화 제목 list
countList = [] # 순위 list

for list in li_list:
    count += 1
    print(count,'~', list.string.strip())
    
    countList.append(count)
    titleList.append(list.string.strip())
```

```python
# list 확인
print(countList)
print(titleList)
```

```python
# countList만으로 data frame 생성
df = pd.DataFrame(data=countList, columns=['순위'])
print(df)
```

```python
# 컬럼 추가하기
df['영화제목'] = pd.Series(titleList)
(df)
```

```python
# csv로 저장
df.to_csv('../data/DaumList.csv', encoding='utf-8', index=False) # index=False는 row.names=F와 같다
```

```python
url = 'https://movie.naver.com/movie/sdb/rank/rmovie.naver'
page = urlopen(url)
soup = BS(page, 'html.parser')
```

```python
movies = soup.select('#old_content > table > tbody > tr > td.title > div > a')
```

```python
count = 0
countList = []
titleList = []
for movie in movies:
    count += 1
    print(movie.string.strip())
    countList.append(count)
    titleList.append(movie.string.strip())
```

```python
print(countList)
print(titleList)
```

```python
naver_movie_ranking = pd.DataFrame(data=countList, columns=['Index'])
naver_movie_ranking
```

```python
naver_movie_ranking['Titles'] = pd.Series(titleList)
naver_movie_ranking
```

```python
naver_movie_ranking.to_csv('../data/naverRanking.csv', encoding='utf-8', index=False) # index=False는 row.names=F와 같다
```
