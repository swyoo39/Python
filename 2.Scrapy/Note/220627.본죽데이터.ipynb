```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
```

```python
# Chrome Browser와 Chrome Driver Version 확인하기

chrome_options = webdriver.ChromeOptions()
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=chrome_options)
driver.get('https://www.bonif.co.kr/store/list')
```

```python
# 현재 화면 Capture 하기
driver.save_screenshot('../data/본죽_매장찾기.png')
```

```python
# 시/도 가져오기
# //*[@id="sido"]
```

```python
sido_raw = driver.find_element(By.XPATH, '//*[@id="sido"]')
sido_list = sido_raw.find_elements(By.TAG_NAME, 'option')
sido_names = [option.get_attribute('value') for option in sido_list]
sido_names.remove('')
print(sido_names, len(sido_names))
```

```python
# 조회버튼 click
# //*[@id="storeC01"]/div/a/img
```

```python
xpath = '//*[@id="storeC01"]/div/a/img'
element_find = driver.find_element(By.XPATH, xpath).click()
```

```python
xpath = '//*[@id="my-store-list"]'
shop_raw = driver.find_element(By.XPATH, xpath)
shopName_list = shop_raw.find_elements(By.TAG_NAME, 'a')
shopDetail_list = shop_raw.find_elements(By.TAG_NAME, 'p')

shopName = []
for i in range(len(shopName_list)//4):
    shopName.append(shopName_list[4*i].text.split('\n'))

shopAddress = []
shopTel = []
for shop in shopDetail_list:
    shopAddress.append(shop.text.split('\n')[0])
    shopTel.append(shop.text.split('\n')[-1])
```

```python
print(len(shopName),len(shopDetail_list))
```

```python
# 화면에 그림을 그리는 시간이 필요하다.
import time

shopName = []
shopAddress = []
shopTel = []

for sido in sido_names:
    # 구 선택하기
    element = driver.find_element(By.XPATH, '//*[@id="sido"]')
    element.send_keys(sido)

    time.sleep(1.5)
    
    element_find = driver.find_element(By.XPATH, '//*[@id="storeC01"]/div/a/img').click()

    time.sleep(1)

    xpath = '//*[@id="my-store-list"]'
    shop_raw = driver.find_element(By.XPATH, xpath)
    shopName_list = shop_raw.find_elements(By.TAG_NAME, 'a')
    shopDetail_list = shop_raw.find_elements(By.TAG_NAME, 'p')

    for i in range(len(shopName_list)//4):
        shopName.append(shopName_list[4*i].text.split('\n'))

    for shop in shopDetail_list:
        shopAddress.append(shop.text.split('\n')[0])
        shopTel.append(shop.text.split('\n')[-1])

    time.sleep(1)
```

```python
print(len(shopName), len(shopAddress), len(shopTel))
```

```python
import pandas as pd

# data frame 만들기
bonShop = pd.DataFrame(data=shopName, columns=['지점명'])

# 컬럼 추가하기
bonShop['위치'] = pd.Series(shopAddress)
bonShop['전화'] = pd.Series(shopTel)
(bonShop)
```
