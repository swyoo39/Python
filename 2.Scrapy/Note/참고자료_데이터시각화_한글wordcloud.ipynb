```python
import nltk # natural language tool kit
from konlpy.corpus import kolaw
from konlpy.tag import Okt
import matplotlib
import matplotlib.pyplot as plt
from wordcloud import WordCloud
t=Okt()
```

```python
# 대한민국 헌법 불러오기
ko_con_text = kolaw.open('constitution.txt').read()
ko_con_text
```

```python
# 명사분리
tokens_ko = t.nouns(ko_con_text)
tokens_ko
```

```python
#불용어 추가
stop_words = ['제','월','일','조','수','때','그','이','바','및','안','정']
tokens_ko = [each_word for each_word in tokens_ko if each_word not in stop_words]
tokens_ko
```

```python
# 상위 50개 단어 확인
ko = nltk.Text(tokens_ko)
ko.vocab().most_common(50)
```

```python
# 한글 폰트 문제 해결 
# matplotlib은 한글 폰트를 지원하지 않음
# os정보
import platform

# font_manager : 폰트 관리 모듈
# rc : 폰트 변경 모듈
from matplotlib import font_manager, rc
# unicode 설정
plt.rcParams['axes.unicode_minus'] = False

if platform.system() == 'Darwin':
    rc('font', family='AppleGothic') # os가 macos
elif platform.system() == 'Windows':
    path = 'c:/Windows/Fonts/malgun.ttf' # os가 windows
    font_name = font_manager.FontProperties(fname=path).get_name()
    rc('font', family=font_name)
else:
    print("Unknown System")
```

```python
# 상위 50개의 빈도수 차트 만들기
plt.figure(figsize=(12,6))
ko.plot(50)
plt.show()
```

```python
# 상위 500개만 word cloud로 그리기
data = ko.vocab().most_common(500)
tmp_data = dict(data)
wordcloud = WordCloud(background_color='White',
                     #font_path = 'c:/Windows/Fonts/malgun.ttf',
                      font_path = 'AppleGothic',
                     relative_scaling = 0.2).generate_from_frequencies(tmp_data)
```

```python
plt.figure(figsize=(16,8))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
from PIL import Image
import numpy as np
korea_coloring = np.array(Image.open('../Data/KoreaArea.jpg'))

from wordcloud import ImageColorGenerator
image_colors = ImageColorGenerator(korea_coloring)
```

```python
plt.figure(figsize=(12, 12))
plt.imshow(korea_coloring, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
wordcloud = WordCloud(background_color='White',
                     #font_path = 'c:/Windows/Fonts/malgun.ttf',
                      font_path = 'AppleGothic',
                     relative_scaling = 0.2,
                     mask=korea_coloring,
                     min_font_size = 1,
                     max_font_size = 40).generate_from_frequencies(tmp_data)
plt.figure(figsize=(12,12))
plt.imshow(wordcloud.recolor(color_func=image_colors), interpolation='bilinear')
plt.axis('off')
plt.show()
```
