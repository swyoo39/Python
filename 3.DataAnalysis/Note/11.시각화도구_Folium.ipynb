### 지도에 마커 표시하기
- 서울 시내 주요 대학교의 위치 표시하기

```python
df = pd.read_excel('../data/서울지역 대학교 위치.xlsx')
df.columns=['학교명', '위도', '경도']
df.head()
```

```python
# 서울 지도 만들기
seoul_map = folium.Map(
    location=[37.55, 126.98], 
    tiles='Stamen Terrain',
    zoom_start=12
)

# 대학교 위치 정보를 Market으로 표시
for name, lat, lng in zip(df.학교명, df.위도, df.경도):
    # print(name, lat, lng)
    folium.Marker([lat,lng], popup=name).add_to(seoul_map)
    
seoul_map
```

---
### 지도에 원형 마커 표시

```python
# 서울 지도 만들기
seoul_map = folium.Map(
    location=[37.55, 126.98], 
    tiles='Stamen Terrain',
    zoom_start=12
)

# 대학교 위치 정보를 Circle Marker로 표시
for name, lat, lng in zip(df.학교명, df.위도, df.경도):
    # print(name, lat, lng)
    folium.CircleMarker(
        [lat,lng], 
        radius=10,
        color='brown',
        # fill=True, # fill_값들이 있으면 자동 True
        # fill=False, # default는 False
        fill_color='coral',
        fill_opacity=0.7,
        popup=name,
    ).add_to(seoul_map)
    
seoul_map
```

### 지도 영역에 단계구분포(Choropleth Map) 표시하기
: 행정구역과 같은 지도상의 어떤 경계에 둘러쌓인 영역에 색을 칠하거나 음영등의 정보를 나타내는 시각화방법

```python
import json
```

```python
# 경기도 인구변화 데이터를 불러와서 dataFrame으로 변환
df = pd.read_excel('../data/경기도인구데이터.xlsx', index_col='구분')
df.head()
```

```python
df.columns = df.columns.map(str) # str값으로 전환
```

```python
# 경기도 시군구 경계정보를 가진 geo-json 파일 불러오기
geo_path = '../data/경기도행정구역경계.json'
try:
    geo_data = json.load(open(geo_path, encoding='utf-8'))
except:
    geo_data = json.load(open(geo_path, encoding='utf-8-sig'))
geo_data
```

```python
# 경기도 지도 만들기
g_map = folium.Map(
    location=[37.5502, 126.982],
    tiles='Stamen Terrain',
    zoom_start=9,
)
# year = '2007'
for year in df.columns:
    # Choropleth 클래스로 단계구분 표시하기
    folium.Choropleth(
        geo_data = geo_data, # 지도 경계
        data=df[year], # 표시하려는 데이터
        columns=[df.index, df[year]], # 열 지정
        fill_color='YlOrRd',
        fill_opacity=0.7,
        line_opacity=0.3,
        threshold_scale=[10000, 100000, 300000, 500000, 700000], # 범위 정하기
        key_on='feature.properties.name', # json파일에서 

    ).add_to(g_map)
    g_map.save('../data/gyonggi_population_'+year+'.html')
    g_map
```
