### CCTV 현황

```python
# CCTV에서 소계로 오름차순 정렬
CCTV_Seoul.sort_values('소계').head()
```

> CCTV가 적은 지역은 도봉구, 중랑구, 강북구, 종로구, 중구 순이다.

```python
# CCTV에서 소계로 오름차순 정렬
CCTV_Seoul.sort_values('소계', ascending=False).head()
```

> CCTV가 많은 지역은 강남구, 관악구, 구로구, 성북구, 은평구 순이다.

```python
#### 최근 3년간 CCTV 증가율을 구해서 CCTV가 제일 많이 증가한 자치구를 검색하고 정렬하기
CCTV_Seoul['최근증가율'] = round((CCTV_Seoul.iloc[:,7:10].sum(axis=1) / CCTV_Seoul.iloc[:,2:7].sum(axis=1)) * 100, 2)
CCTV_Seoul.sort_values('최근증가율', ascending=False).head()
```

> 결과를 보면 최근 3 년간의 CCTV가 2015 년 이전에 비해 가장 많이 증가한 곳은 중구, 영등포구, 성동구, 금천구, 광진구 순이다.

### 인구현황

```python
# 구별 Data의 유효성 확인 : 자치구 25개
len(pop_seoul1['구별'].unique())
```

```python
pop_seoul1.head()
```

```python
pop_seoul1['외국인비율'] = round(pop_seoul1.외국인 / pop_seoul1.인구수 * 100, 2)
pop_seoul1['고령자비율'] = round(pop_seoul1.고령자 / pop_seoul1.인구수 * 100, 2)
pop_seoul1.head()
```

```python
# 인구수가 제일 많은 지역 top5
pop_seoul1.sort_values('인구수', ascending=False).head()
```

```python
# 외국인이 제일 많은 지역 top5
pop_seoul1.sort_values('외국인', ascending=False).head()
```

> 외국인이 많은 지역은 영등포구+구로구+금천구로 밀집되어있는 경향을 보인다.

```python
# 외국인비율이 제일 높은 지역 top5
pop_seoul1.sort_values('외국인비율', ascending=False).head()
```

> 외국인비율이 높은 지역은 중구+용산구 그룹, 영등포구+구로구+금천구 그룹으로 비교적 비슷한 지역에 몰려있다.

```python
# 고령자 top5
pop_seoul1.sort_values('고령자', ascending=False).head()
```

> 고령자 수가 많은 지역은 서울 경계(외곽) 지역에 분포해있다.

```python
# 고령자비율 top5
pop_seoul1.sort_values('고령자비율', ascending=False).head()
```

> 고령자비율이 높은 지역은 전부 강북이며, 종로구 성북구를 중심으로 많이 밀집되어있는 경향이 있다.

```python
pop_seoul1.corr()
```

- 인구수가 제일 많은 지역은 송파구지만, 영등포구가 외국인이 제일 많다.   
- 그리고 그 근처의 구로구, 금천구 등도 외국인이 많다.   
- 송파구는 고령자가 제일 많지만, 고령자비율은 높지 않다.   
- 하지만 강북구, 도봉구, 중구, 종로구, 은평구 같은 경우는 인구수 대비 고령자비율이 높기 때문에 추후 '고령화'가 우려된다.

```python
# 인구수, 소계에 따른 기준직선(회귀선, 평균선) 만들기
fp1 = np.polyfit(data_result.인구수, data_result.소계, 1)
fp1
```

```python
# x축 만들기 : 인구수 표현
fx = np.linspace(100000, 700000)

# y축 만들기 : 소계(CCTV) 표현
f1 = np.poly1d(fp1) # 1차 함수

print(fx)
print(f1)
```

```python
plt.figure(figsize=(10, 10))

plt.scatter(data_result.인구수, data_result['소계'])
plt.plot(
    fx, 
    f1(fx), 
    ls='dashed', # line style
    lw=3, # line width
    color='r'
    )

plt.title('인구수, 소계에 따른 기준직선 만들기', size=25)
plt.xlabel('인구수', size=15, color='blue')
plt.ylabel('CCTV 소계', size=15, color='blue')
plt.grid()

plt.show()
```

> 이 직선은 cctv 소계의 기준값이다.   
> 이 직선을 기준으로 상위에 있는 점은 많이 설치된 것   
> 하위에 있는 점은 적게 설치된 것

```python
# 기준선 구하기
fp1 = np.polyfit(data_result.인구수, data_result.소계, 1)

# 기준치 구하기
f1 = np.poly1d(fp1)

# 인구수
# fx = np.linspace(100000, 700000)

# 소계에서 기준치를 빼서 오차값 구하기
data_result['오차'] = np.abs(data_result.소계-f1(data_result.인구수))

# 정렬
df_sort = data_result.sort_values('오차', ascending=False)
df_sort.head()
```

```python
# 그래프에 텍스트와 color를 기준선 기준으로 입히기
plt.figure(figsize=(14, 10))

plt.scatter(
    data_result.인구수, 
    data_result.소계, 
    c=data_result['오차'], # 오차를 기준으로 색을 입히기
    )
plt.plot(
    fx, 
    f1(fx), 
    ls='dashed', # line style
    lw=3, # line width
    color='r'
    )

# 오차가 가장 큰 10개 구에 텍스트 입히기
for n in range(10):
    plt.text(df_sort.인구수[n]*1.02, df_sort.소계[n]*0.98, df_sort.index[n], fontsize=15)

plt.title('인구수, 소계에 따른 기준직선 만들기', size=25)
plt.xlabel('인구수', size=15, color='blue')
plt.ylabel('CCTV 소계', size=15, color='blue')
plt.colorbar()
plt.grid()

plt.show()
```
