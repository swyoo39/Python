```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("../Data/시도별 전출입 인구수.xlsx")
df.head()
```

```python
df.tail()
```

```python
# NaN을 전(Prev) 데이터로 채움 / 근처에 있는걸로 채움
# NaN 값을 전국으로 바꾸기
df = df.fillna(method='ffill')
df.head()
```

```python
df.info()
```

```python
df[df['전입지별'].isnull()]
```

```python
# 서울에서 다른 지역으로 이동한 데이터만 추출하여 정리
# 전출지별이 서울이고 전입지별이 서울이 아닌 데이터

df_seoul = df[(df['전출지별'] == '서울특별시') & (df['전입지별'] != '서울특별시')]
df_seoul.head(10)
```

```python
# 전출지별 칼럼 삭제

df_seoul.drop(['전출지별'],axis=1, inplace=True)
df_seoul.head()
```

```python
# 전입지별 ->전입지로 컬럼 이름 변경
df_seoul.rename(columns={'전입지별':'전입지'}, inplace=True)
```

```python
df_seoul.head()
```

```python
# df_seoul의 index를 전입지로 변경
df_seoul.set_index('전입지', inplace=True)
```

```python
df_seoul.head()
```

```python
# 서울에서 경기도로 이동한 인구 값만 선택
sr_one = df_seoul.loc['경기도']
sr_one.head()
```

```python
plt.plot(sr_one.index, sr_one.values)
plt.show()
```

```python
plt.plot(sr_one) # rownames가 x로, data들이 y로
```

```python
# 한글 폰트 문제 해결 
# matplotlib은 한글 폰트를 지원하지 않음
# os정보
import platform

# font_manager : 폰트 관리 모듈
# rc : 폰트 변경 모듈
from matplotlib import font_manager, rc
# unicode 설정
plt.rcParams['axes.unicode_minus'] = False

if platform.system() == 'Darwin':
    rc('font', family='AppleGothic') # os가 macos
elif platform.system() == 'Windows':
    path = 'c:/Windows/Fonts/malgun.ttf' # os가 windows
    font_name = font_manager.FontProperties(fname=path).get_name()
    rc('font', family=font_name)
else:
    print("Unknown System")
```

```python
# 차트 제목 추가
plt.plot(sr_one)
plt.title('서울시 -> 경기도 인구이동')
# plt.show() # [<~>] 내용 안보인다.


# 축 이름 추가
plt.xlabel('기간')
plt.ylabel('이동인구 수')
plt.show()
```

```python
# 그래프 꾸미기
# 그림 사이즈 지정(가로 14인치, 세로 5인치)
plt.figure(figsize=(14,5))
# x축 라벨 회전하기
plt.xticks(rotation='vertical') # 숫자 적으면 해당 각도만큼 회전
# title, labels
plt.title('서울 -> 경기 인구이동')
plt.xlabel('기간')
plt.ylabel('이동인구 수')
plt.plot(sr_one)
# 범례 표시
plt.legend(labels = ['서울 -> 경기'], loc = 'best') # 알아서 제일 좋은 곳을 찾아준다.
plt.show()
```

### Style 추가해보기

```python
# Style 서식 지정
plt.style.use('ggplot')
# 그림 사이즈 지정(가로 14인치, 세로 5인치)
plt.figure(figsize=(14,7))
# Marker 표시
plt.plot(sr_one, marker='o', markersize=10)
# x축 라벨 회전하기
plt.xticks(rotation='vertical') # 숫자 적으면 해당 각도만큼 회전
# title, labels
plt.title('서울 -> 경기 인구이동', size=20, )
plt.xlabel('기간', size=15)
plt.ylabel('이동인구 수', size=15)
# y축 범위 지정
plt.ylim(50000, 700000) # ylim(최소값, 최대값)
# 화살표 표시
plt.annotate('',
    xy = (20,620000), # 화살표의 머리부분
    xytext=(2,290000), # 화살표의 꼬리부분(글자 써지는 위치)
    # xycoords='data', # 좌표체계(data type), 현재는 필요없다.
    arrowprops=dict(arrowstyle='->', color='skyblue', lw=5),
    )
# 글자 쓰기
plt.annotate(
    '인구이동 증가(1970~1995)', # 내용
    xy = (10, 400000), # 위치 기준점
    rotation = 36, # 회전 각도
    va = 'baseline', # 상하 정렬
    ha = 'center', # 좌우 정렬
    fontsize = 12, # 글씨 크기
)
# 화살표 표시 2
plt.annotate('',
    xy = (47,400000), # 화살표의 머리부분
    xytext=(30,600000), # 화살표의 꼬리부분(글자 써지는 위치)
    # xycoords='data', # 좌표체계(data type), 현재는 필요없다.
    arrowprops=dict(arrowstyle='->', color='olive', lw=5),
    )
# 글자 쓰기 2
plt.annotate(
    '인구이동 감소(1995~2017)', # 내용
    xy = (39, 470000), # 위치 기준점
    rotation = -24.5, # 회전 각도
    va = 'baseline', # 상하 정렬
    ha = 'center', # 좌우 정렬
    fontsize = 12, # 글씨 크기
)
# 범례 표시
plt.legend(labels = ['서울 -> 경기'], loc = 'best', fontsize=15) # 알아서 제일 좋은 곳을 찾아준다.
# 이미지 저장
plt.savefig('../data/seoul.png')
plt.show()
```

### 화면을 분할하여 그래프 여러개 그리기

```python
# Warning 무시
import warnings
warnings.filterwarnings('ignore')
```

```python
# Style 서식 지정
plt.style.use('ggplot')
# 그래프 객체 생성(figure에 두개의 서브플롯 생성)
fig = plt.figure(figsize=(10,14))
ax1 = fig.add_subplot(2,1,1) # 2행 1열의 첫번째다.
ax2 = fig.add_subplot(2,1,2) # 2행 1열의 두번째다.

ax1.plot(sr_one, '*', markersize=10)
ax2.plot(
    sr_one, 
    marker='o', # 선 + 도형 그래프 : o, *, +, |, ., ^
    markerfacecolor='green', # 도형 안 색
    markersize=10, # 도형 크기
    color='olive', # 선 색
    linewidth=2, # 선 두께
    label='서울->경기' # legend
    )
ax2.legend(loc='best')

ax2.set_title('서울 ~ 경기 인구이동', fontsize=20)

ax1.set_ylim(50000,700000)
ax2.set_ylim(50000,700000)

ax1.set_xticklabels(sr_one.index, rotation=90)
ax2.set_xticklabels(sr_one.index, rotation=90)

plt.show()
```

### 동일한 그림에 여러개의 그래프를 추가

```python
# 서울에서 충청남도, 경상북도, 강원도로 이동한 인구만 선택하기
df_3 = df_seoul.loc[['충청남도', '경상북도', '강원도']]
df_3
```

```python
# Style
plt.style.use('ggplot')

fig = plt.figure(figsize=(20,5))

ax = fig.add_subplot(1,1,1)

ax.plot(
    df_3.columns, df_3.loc['충청남도'],
    marker = 'o',
    markerfacecolor = 'green',
    markersize = 10,
    color = 'olive',
    linewidth = 2,
    label = '서울 -> 충남'
    )

ax.plot(
    df_3.columns, df_3.loc['경상북도'],
    marker = 'o',
    markerfacecolor = 'blue',
    markersize = 10,
    color = 'skyblue',
    linewidth = 2,
    label = '서울 -> 경북'
    )

ax.plot(
    df_3.columns, df_3.loc['강원도'],
    marker = 'o',
    markerfacecolor = 'red',
    markersize = 10,
    color = 'magenta',
    linewidth = 2,
    label = '서울 -> 강원도'
    )
ax.legend(loc='best')

ax.set_title('서울 -> 충남, 경북, 강원 인구이동', size = 20)
ax.set_xlabel('기간', size = 15)
ax.set_ylabel('이동인구', size = 15)

ax.set_xticklabels(df_3.columns, rotation=90)

# 축 라벨 크기
ax.tick_params(axis='x', labelsize=9)
ax.tick_params(axis='y', labelsize=9)
plt.show()
```

```python
# 서울에서 충청남도, 경상북도, 강원도로 이동한 인구만 선택하기
df_4 = df_seoul.loc[['충청남도', '경상북도', '강원도', '전라남도']]
df_4
```

```python
# Style 서식 지정
plt.style.use('ggplot')
# 그래프 객체 생성(figure에 두개의 서브플롯 생성)
fig = plt.figure(figsize=(20,10))
ax1 = fig.add_subplot(2,2,1) # 2행 2열의 첫번째다.
ax2 = fig.add_subplot(2,2,2) # 2행 2열의 두번째다.
ax3 = fig.add_subplot(2,2,3) # 2행 2열의 세번째다.
ax4 = fig.add_subplot(2,2,4) # 2행 2열의 네번째다.

ax1.plot(
    df_3.columns, df_3.loc['충청남도'],
    marker = 'o',
    markerfacecolor = 'green',
    markersize = 10,
    color = 'olive',
    linewidth = 2,
    label = '서울 -> 충남'
    )

ax2.plot(
    df_3.columns, df_3.loc['경상북도'],
    marker = 'o',
    markerfacecolor = 'blue',
    markersize = 10,
    color = 'skyblue',
    linewidth = 2,
    label = '서울 -> 경북'
    )

ax3.plot(
    df_3.columns, df_3.loc['강원도'],
    marker = 'o',
    markerfacecolor = 'red',
    markersize = 10,
    color = 'magenta',
    linewidth = 2,
    label = '서울 -> 강원도'
    )

ax4.plot(
    df_4.columns, df_4.loc['전라남도'],
    marker = 'o',
    markerfacecolor = 'orange',
    markersize = 10,
    color = 'yellow',
    linewidth = 2,
    label = '서울 -> 전남'
    )

ax1.legend(loc='best')
ax2.legend(loc='best')
ax3.legend(loc='best')
ax4.legend(loc='best')

ax1.set_title('서울 -> 충남 인구이동', size = 15)
ax2.set_title('서울 -> 경북 인구이동', size = 15)
ax3.set_title('서울 -> 강원 인구이동', size = 15)
ax4.set_title('서울 -> 전남 인구이동', size = 15)
# ax.set_xlabel('기간', size = 15)
# ax.set_ylabel('이동인구', size = 15)

ax1.set_ylim(5000, 65000)
ax2.set_ylim(5000, 65000)
ax3.set_ylim(5000, 65000)
ax4.set_ylim(5000, 65000)

ax1.set_xticklabels(df_4.columns, rotation=60)
ax2.set_xticklabels(df_4.columns, rotation=60)
ax3.set_xticklabels(df_4.columns, rotation=60)
ax4.set_xticklabels(df_4.columns, rotation=60)

ax1.tick_params(axis='x', labelsize=8)
ax2.tick_params(axis='x', labelsize=8)
ax3.tick_params(axis='x', labelsize=8)
ax4.tick_params(axis='x', labelsize=8)

plt.show()
```

### 2010~2017년의 기간동안 서울에서 각 시도로 이동한 인구의 합계를 구하여 그래프로 비교

```python
# 서울에서 충남, 경북, 강원, 전남으로 이동한 인구 데이터만 선택
col_years = list(map(str, range(2010, 2018)))
col_years
```

```python
df_4_2010 = df_seoul.loc[['충청남도', '경상북도', '강원도', '전라남도'], col_years]
df_4_2010
```

```python
df_4_2010['합계'] = df_4_2010.sum(axis=1)
df_4_2010
```

```python
df_total = df_4_2010[['합계']].sort_values(by='합계', ascending=True)
df_total
```

```python
plt.style.use('ggplot')

df_total.plot(
    kind='barh',
    color='cornflowerblue',
    figsize=(10,5)
    )

plt.title('서울 -> 타시도 인구이동')
plt.ylabel('전입지')
plt.xlabel('이동인구')

plt.show()
```
