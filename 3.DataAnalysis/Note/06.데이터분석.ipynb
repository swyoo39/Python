```python
train.loc[train['Nationality_AT'],'Nationality'] = 'Austria'
train.loc[train['Nationality_UK'],'Nationality'] = 'TheUnitedKingdom'
```

```python
# Fare
train['Fare'].describe()
```

```python
# 15보다 작으면 LowFare, 15~30 은 MediumFare, 30보다 크면 HighFare
train['LowFare'] = train.Fare < 15
train['MediumFare'] = (train.Fare >= 15) & (train.Fare < 30)
train['HighFare'] = train.Fare >= 30
```

```python
train[['Fare','LowFare','MediumFare','HighFare']].head(10)
```

```python
train.loc[train.LowFare,'FareType'] = 'Low'
train.loc[train.MediumFare,'FareType'] = 'Medium'
train.loc[train.HighFare,'FareType'] = 'High'
```

```python
train[['Fare','FareType']].head(10)
```

```python
def check_fare_type(fare):
    if fare < 15:
        return 'Low'
    elif fare >= 15 and fare < 30:
        return 'Medium'
    else:
        return 'High'
```

```python
train['FareType'] = train.Fare.apply(check_fare_type)
train[['Fare','FareType']].head(10)
```

```python
train.Age.replace()
```

```python
# Age 의 NaN을 Age의 평균값으로 대체하기 (평균나이는 소수점 첫째자리까지)
train.loc[train.Age.isnull(),'Age'] = round(train.Age.mean(), 1)
train['Age'].head(10)
```

```python
train.Age = round(train.Age, 1)
```
