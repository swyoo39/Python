# 데이터 사전처리(Preprocessing)
- 머신러닝 등 데이터 분석의 정확도는 분석 데이터의 품질에 좌우된다.
- 데이터의 품질을 높이기 위해서 누락, 중복 데이터 등의 오류를 수정하고 분석 목적에 맞게 변형하는 과정이 필요하다.
- 수집한 데이터를 분석 목적에 적합하게 하는 작업을 사전처리(Preprocessing)이라고 한다.

### 누락 데이터(NaN : Not a Number)
- 데이터를 파일로 입력할 때나, 파일 형식을 변환하는 과정에서 데이터가 손실되는 경우가 발생
- 머신러닝 분석 모형에 데이터를 입력하기 전에 반드시 누락 데이터를 제거하고 다른 값으로 대체해야 한다.
- 누락 데이터가 많아지면 데이터의 품질이 떨어지고, 머신러닝 분석 알고리즘을 왜곡하는 현상이 발생한다.

# 데이터 표준화
- 여러 곳에서 수집한 데이터는 서로 다른 단위가 섞이거나, 대상을 다른 형식으로 표현하는 경우가 많다.
- 대상을 표현하는 방법에 차이가 있으면 분석의 정확도는 현저히 낮아진다.
- 데이터 포맷을 일관성 있게 하는 작업이 표준화이다.

> mpg : mile per gallon ~ 한국과 단위가 다르다.

# 단위 변환
mpg(mile per gallon) >> kpl(kilometer per liter)
- 1 mile = 1.60934 km
- 1 gallon = 3.78541 liter

```python
mpg_to_kpl = 1.60934 / 3.78541
df['kpl'] = round(df.mpg * mpg_to_kpl, 2)
df.head()
```

### 범주형(카테고리) 데이터 처리
- 연속 데이터를 그대로 사용하기보다는 일정한 구간(binning)으로 나눠서 분석하는 것이 효율적인 경우가 많다.

```python
# horsepower를 3개의 범위로 나누기
count, bin_dividers = np.histogram(df['horsepower'], bins=3)
count, bin_dividers
```

```python
bin_dividers
```

> 257, 103, 32 부분은 count값, 46~107/107~168/168~230 부분은 범위를 나눠준 것.

```python
# 3개의 bin에 이름 지정
bin_names = ['저출력','보통출력','고출력']
```

```python
# cut by bin
df['hp_bin'] = pd.cut(
    x = df['horsepower'],
    bins=bin_dividers,
    labels=bin_names,
    include_lowest=True # 첫 경계값 포함
)
```

```python
df[['horsepower', 'hp_bin']].head(15)
```

> 머신러닝에서 범주형의 문자열을 쓸 수 없다!

### One-Hot Encoding
- 범주형 데이터를 0,1 을 갖는 원핫벡터로 변환
- 결과는 선형대수학에서 정의하는 희소행렬(sparse matrix)로 변환
- 희소행렬은 행/열 좌표와 값 형태로 정리

```python
from sklearn import preprocessing # 사이키런
```

```python
# 전처리를 위한 encoder 객체 만들기
label_encoder = preprocessing.LabelEncoder() # label encoder 생성
onehot_encoder = preprocessing.OneHotEncoder()
```

```python
df['hp_bin'].head(15)
```

```python
# label encoder로 문자열 범주를 숫자형 범주로 바꾸기
onehot_labeled=label_encoder.fit_transform(df['hp_bin'].head(15))
onehot_labeled
```

> 보통출력 : 1, 고출력 : 0, 저출력 : 2

```python
# 2차원 행렬로 형태를 변경하자 ~ 희소행렬의 좌표값으로 쓰기위해
onehot_reshaped = onehot_labeled.reshape(len(onehot_labeled),1)
onehot_reshaped
```

```python
# 희소행렬
onehot_fitted = onehot_encoder.fit_transform(onehot_reshaped)
print(onehot_fitted)
```

> 각 좌표에 1 값이 들어있다.    
> 각 좌표의 열 값이 onehot_labeled에 저장된 값으로 되어있다.

### 정규화(Normalization)
- 각 열에 있는 숫자 데이터의 상대적 크기 때문에, 머신러닝 분석결과가 달라질 수 있다.
- 숫자 데이터의 상대적인 크기 차이를 제거

```python
# 최대값의 절대값을 이용한 정규화
df.horsepower.describe()
```

```python
# horsepower 열의 최대값의 절대값(abs)으로 모든 데이터를 나누기 : 0과 1 사이의 값으로 변환
horsepower1 = df.horsepower / abs(df.horsepower.max())
horsepower1.describe()
```

```python
df.horsepower.describe()
```

```python
# 각 열의 데이터에서 최소값을 뺀 값을 분자로
min_x = df.horsepower - df.horsepower.min()

# 각 열의 최대값에서 최소값을 뺀 값을 분모로
min_max = df.horsepower.max() - df.horsepower.min()

# min_x / min_max 는 0과 1 사이의 값이 나오게 된다.
horsepower2 = min_x / min_max
horsepower2.describe()
```

### Timestamp를 Period로 변환

```python
# 날짜 형식의 문자열로 구성되는 리스트 정의
dates = ['2019-01-01', '2020-03-01', '2021-06-01']
```

```python
ts_dates = pd.to_datetime(dates)
ts_dates
```

```python
pr_day = ts_dates.to_period(freq='D')
print(pr_day)
pr_month = ts_dates.to_period(freq='M')
print(pr_month)
pr_year = ts_dates.to_period(freq='Y') # Y 혹은 A
print(pr_year)
```

> freq Option   
> D (Day: 1일)   
> W (Week:1주)   
> M (Month end: 월말)   
> MS (Month begin: 월초)   
> Q (Quarter end : 분기말)   
> QS (Quarter begin: 분기초)   
> A (year end : 연말)   
> AS (year begin: 연초)   
> B (Business day : 휴일 제외)   
> H (hour : 1시간)   
> T (minute :1분)   
> S (second: 1초)

### 시계열 데이터 만들기

```python
# Timestamp의 배열 만들기 : 월 간격, 월 시작일 기준
ts_ms = pd.date_range(
    start='2022-01-01', # 날짜 범위의 시작
    end=None, # 날짜 범위의 끝(미정)
    periods=6, # 생성할 Timestamp의 개수
    freq='MS', # 시간 간격 : 월초
    tz='Asia/Seoul', # timezone 시간대
)
ts_ms
```

```python
# 월 간격, 월 마지막날 기준
ts_me = pd.date_range(
    start='2022-01-01', # 날짜 범위의 시작
    end=None, # 날짜 범위의 끝(미정)
    periods=6, # 생성할 Timestamp의 개수
    freq='M', # 시간 간격 : 월말
    tz='Asia/Seoul', # timezone 시간대
)
ts_me
```

```python
# 분기(3개월) 간격, 월의 마지막날 기준
ts_3m = pd.date_range(
    start='2022-01-01', # 날짜 범위의 시작
    end=None, # 날짜 범위의 끝(미정)
    periods=6, # 생성할 Timestamp의 개수
    freq='3M', # 시간 간격 : 3개월 말
    tz='Asia/Seoul', # timezone 시간대
)
ts_3m
```
