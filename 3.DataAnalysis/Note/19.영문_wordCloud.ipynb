```python
import nltk
import pickle
from nltk.corpus import stopwords # 문장 하나 corpus, 불용어 stopwords
import re # 정규식
nltk.download('all')
```

## text를 불러서 word 분류

```python
content = open('../data/Independence.txt').read()
content
```

```python
# 불필요한 심벌 없애기
# cleaned_content = content.replace('!', '').replace(',','') # 이렇게도 가능하다
cleaned_content = re.sub(r'[^\.\?\!\w\d\s]', '', content) # 문장단위로 끊기
cleaned_content
```

```python
cleaned_content = cleaned_content.lower()
cleaned_content
```

```python
# 문장에서 워드를 토큰(단어)으로 쪼갠다.
word_tokens = nltk.word_tokenize(cleaned_content)
print(word_tokens)
```

```python
# 품사 분리하기
token_pos = nltk.pos_tag(word_tokens)
print(token_pos)
```

```python
# 명사만 추출하기
NN_words = []

for word, pos in token_pos:
    if 'NN' in pos: # 명사
        NN_words.append(word)

print(NN_words) # 복수형도 포함됐다.
```

```python
# 원형으로 복원
wlem = nltk.WordNetLemmatizer()
lemmatized_words = []

for word in NN_words:
    new_word = wlem.lemmatize(word)
    lemmatized_words.append(new_word)

print(lemmatized_words)
```

```python
# 불용어 제거
stopwrods_list = stopwords.words('english')
stopwrods_list
unique_NN_Words = set(lemmatized_words)
final_NN_words = lemmatized_words

for word in unique_NN_Words:
    if word in stopwrods_list: # set된 단어들로 제거할 불용어 찾고
        while word in final_NN_words: # 중복되는 불용어들 다 지우기 위한 while
            final_NN_words.remove(word)
```

```python
print(final_NN_words)
```

```python
# 원하는 단어 제거
customized_stopwords = ['be', 'today', 'yesterday', "it's", "don't"]

unique_NN_Words_1 = set(final_NN_words)

for word in unique_NN_Words_1:
    if word in customized_stopwords:
        while word in final_NN_words:
            final_NN_words.remove(word)
```

> 띄워보고 필요없는 단어를 발견했을 때 시도

```python
# 빈도분석
from collections import Counter

c = Counter(final_NN_words)
print(c)
```

```python
print(c.most_common(20)) # 제일 빈도수가 높은 20개 단어만 순서대로 추출
```

```python
# 워드 클라우드
import wordcloud
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
noun_text = ''
for word in final_NN_words:
    noun_text = noun_text + ' ' + word
    
noun_text
```

```python
wordcloud = WordCloud(
    max_font_size=60,
    relative_scaling=0.5,
).generate(noun_text) # generate가 noun_text의 string value를 입력받아 count 해준다.

plt.figure(figsize=(12, 10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')

plt.show()
```
