# 머신러닝이란?
: 수 많은 데이터를 학습시켜 그 패턴(규칙, 알고리즘, 선형회귀식, a/b 값)을 찾아내고 이를 기반으로 데이터를 분류하거나 예측.   
- 일반 소프트웨어(결정론적 소프트웨어)는 언제나 똑같은 과정을 거쳐 똑같은 결과를 산출한다.   
- 머신러닝 기반 소프트웨어(비결정론적 소프트웨어)는 데이터와 알고리즘에 따라 때로는 다른 결과를 추론한다.   
(추론 : 이미 알고있는 것으로부터 논리적 결론을 도출하는 과정.)

### 결정론적 소프트웨어와 비결정론적 소프트웨어의 차이점
: 예를 들어, 지문인식기의 소프트웨어를 결정론적 소프트웨어로 처리했다고 하면, 지문이 조금만 틀어져도 거부메세지가 나온다.   
: 이에 반해 비결정론적 소프트웨어는 지문이 조금 틀어져도 추론을 통해 지문의 가부를 결정한다.

### 머신러닝 개요
: 머신러닝이란 인공지능 연구과제 중의 하나로, 인간의 뇌가 자연스럽게 수행하는 '학습'이란 능력을 컴퓨터로 구현하는 방법이다.   
- 어느정도 규모가 있는 샘플 데이터를 입력하여 분석시키고, 그 데이터에서 일정한 규칙을 찾아낸다.
- 그리고 찾아낸 규칙을 기반으로 다른 데이터를 분류하거나(classification) 예측(regression)하는 것.

> 학습시키는 방법에 따라 SVM, knn, 선형회귀 등이 있다.

> train 데이터만 맞추고 test 데이터는 못 맞추는, 과적합

### 머신러닝의 종류
#### 지도학습(Supervised Learning)
: 학습 시 데이터와 답을 같이 학습한 후, 다른 데이터의 값을 예측 > 데이터가 많아야 유의미하다.
#### 비지도학습(Unsupervised Learning)
: 학습 시 데이터만 학습한 후, 다른 데이터의 규칙성을 찾음 > 본질적인 구조 등을 확인할 때 사용한다.
#### 강화학습(Reinforcement Learning)
: 부분적으로 답을 입력한 후, 데이터를 기반으로 최적의 답을 찾음 > 알파고, 강화학습~딥러닝
#### 딥러닝(Deep Learning)
: 지도, 비지도 학습과 관련이 있으나 "학습자료에 대한 인간의 개입을 받지 않음"

### 머신러닝의 흐름
(1) 데이터 수집   
(2) 데이터 가공   
(3) 데이터 학습   
&nbsp; (3-1) 학습 방법 선택   
&nbsp; (3-2) 매개 변수 조정   
&nbsp; (3-3) 모델 학습   
(4) 모델 평가   
(5) 예측력이 좋은 경우 진행, 아니면 (3)번으로   
(6) 성공   

## 머신러닝 응용분야
### 클래스 분류(Classification)
- 특정 데이터에 레이블(label, 정답 값)을 붙여 분리할 수 있다.
- 스팸메일 분류, 필기 인식, 증권사기 방지

### 클러스터링(Clustering) : grouping
- 값의 유사성을 기반으로 데이터를 여러 그룹으로 나눌 수 있습니다.
- 사용자의 취향을 그룹으로 묶어 사용자 취향에 맞는 광고를 제공

### 추천(Recommendation)
- 특정 데이터를 기반으로 다른 데이터를 추천하는 것입니다.
- 인터넷 섲머에서 구매한 책들을 기반으로 다른 책을 추천

### 회귀(Regression)
- 과거의 데이터를 기반으로 미래를 예측한 것입니다.
- 판매 예측, 주가 변동

### 차원 축소(Dimensionality Reduction)
- 데이터의 특성을 유지하면서 데이터의 양을 줄이는 것입니다. for performance
- 데이터를 시각화하거나 구조를 추출하여 용량을 줄임으로써 계산을 빠르게 하거나 메모리를 절약할 때 사용

# 프로젝트 과정
: 머신러닝 프로젝트는 크게 '문제 정의, 데이터 획득, 모델 구현, 검증, 테스트'의 단계로 구현

## 문제 정의
- 프로젝트의 목표를 정의하고 그에 따른 계획을 세우는 단계
- 대략적으로 어떤 데이터를 수집해서 어떤 머신러닝 알고리즘으로 문제를 해결할 지 결정하는 단계

## 데이터 획득
- 머신러닝 모델 학습에 필요한 데이터를 수집하는 단계
- 성공적인 머신러닝 프로젝트는 뛰어난 머신러닝 실력보다 얼마나 충분히 데이터를 획득했느냐에 의해 결정이 됩니다.
- 획득한 데이터는 '학습데이터, 검증데이터, 테스트데이터'로 구분되어 사용된다.

## 모델 구현
- 학습데이터를 기반으로 한, 한 개 이상의 머신러닝 모델을 구현

## 검증
- 검증 데이터를 사용해 구현된 머신러닝 모델들의 성능을 검증
- 최종적으로 검증 결과가 가장 높은 모델 하나를 선택

## 테스트
- 검증 결과를 통ㅇ해 선택된 모델의 성능을 테스트데이터로 측정 (일반화 성능 측정)

> 보통 테스트데이터를 20 %로 많이 한다.
